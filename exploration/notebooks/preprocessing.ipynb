```python
import string
import re

HEBREW_LETTERS = ''.join((chr(c) for c in range(ord('\u05D0'), ord('\u05EA') + 1)))
ENGLISH_LETTERS = string.ascii_letters
PUNCT = string.punctuation + " \n"
NUMBERS = string.digits
LETTERS_WHITELIST = ''.join((HEBREW_LETTERS, ENGLISH_LETTERS, PUNCT, NUMBERS))

DATE_REGEXP = r'\d{1,2}[\.\-\/\\]\d{1,2}[\.\-\/\\]\d{2}(\d{2})?'
NUMBER_REGEXP = r'\d+'

DATE_TOKEN = '[DATE]'
NUMBER_TOKEN = '[NUM]'

def clean_text(text):
    # clean characters
    clean_text = ''.join((c for c in text if c in LETTERS_WHITELIST))
    
    # placeholders
    clean_text_with_placeholders = re.sub(DATE_REGEXP, DATE_TOKEN, clean_text)
    clean_text_with_placeholders = re.sub(NUMBER_REGEXP, NUMBER_TOKEN, clean_text_with_placeholders)
    
    return clean_text_with_placeholders
```

```python
import os

FILES = ['all_text.txt', 'test2.txt', 'train2.txt']

for file in FILES:
    file_name = file[:-4] + '.clean.txt'
    
    with open(file) as f_in:
        with open(file_name, 'w') as f_out:
            for line in f_in:
                t = clean_text(line) + '\n'
                f_out.write(t)
```

```python
import nltk

nltk.download('punkt')
```

```python
from tqdm import tqdm

NEW_FILES = ['all_text.clean.txt', 'test2.clean.txt', 'train2.clean.txt']
NEW_FILES = ['train2.clean.txt']

for file in NEW_FILES:
    with open(file, 'r') as f:
        content = f.read().splitlines()
    total = ''
    for line in tqdm(content, desc='line'):
        sublines = nltk.sent_tokenize(line)
        newlines = "\n".join(sublines)
        total += newlines + "\n"
    file_name = file[:-4] + '.lines.txt'
    with open(file_name, 'w') as f_out:
        f_out.write(total)
```
