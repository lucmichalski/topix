```python
from tokenizers import BertWordPieceTokenizer
```

```python
tokenizer = BertWordPieceTokenizer(lowercase=True, vocab_file='./hebrew_vocabs/tokenizer_wp/vocab.txt', add_special_tokens=True, unk_token='[UNK]', 
                                      sep_token='[SEP]', cls_token='[CLS]', clean_text=True, handle_chinese_chars=False, 
                                      strip_accents=True, wordpieces_prefix='##')
```

```python
tokenizer.encode(lines[0])
```

```python
tc = [len(tokenizer.encode(line)) for line in lines]
```

```python
tc[:1000]
```

```python
sum(tc)/len(tc)
```

```python
max(tc)
```

```python
import numpy as np
```

```python
indicies = np.argsort(tc)
indicies = indicies[::-1]
```

```python
for index in indicies[:1000]:
    print(tc[index])
```

```python
normal_paragraphs = [lines[i] for i in range(len(lines)) if tc[i] < 512]
```

```python
with open("./data/all_text_paragraphs_line_by_line.clean.txt", "w") as f:
    for paragraph in normal_paragraphs:
        f.write(paragraph)
        f.write("\n")
```
